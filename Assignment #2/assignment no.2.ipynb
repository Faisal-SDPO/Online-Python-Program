```python
my_list = [10, 20, 30, 40, 50]

length = len(my_list)
print("Length of the list is:", length)
```

```python
my_list = [10, 20, 30, 40, 50]

total = sum(my_list)
print("Sum of all elements:", total)
```

```python
my_list = [10, 25, 5, 80, 45]

largest = max(my_list)
print("Largest number is:", largest)
```
